```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#IMPORTING THE CSV FILE
df=pd.read_csv("student-math.csv",sep=';')
```

```python
df.head()
```

```python
#CREATING A NEW COLUMN FINAL_GRADE
df["final_grade"]=df["G1"]+df["G2"]+df["G3"]
```

```python
df.head()
```

```python
for x in df:
    if df[x].nunique()==2:
        data_dict={df[x].unique()[0]:0,df[x].unique()[1]:1}
        df[x]=df[x].map(data_dict)
```

```python
df.head()
```

```python
df.tail()
```

```python
d=[np.array(df["final_grade"][df["studytime"]==1]),np.array(df["final_grade"][df["studytime"]==2]),
            np.array(df["final_grade"][df["studytime"]==3]),np.array(df["final_grade"][df["studytime"]==4])]
```

```python
fig,axes=plt.subplots(1,2,figsize=(20,8))
axes[0].set_title("STUDYTIME VS. FINALGRADE(SCATTERPLOT)")
axes[1].set_title("STUDYTIME VS. FINALGRADE(BOXPLOT)")
for ax in axes:
    ax.set_xlabel("STUDY TIME")
    ax.set_ylabel("FINAL GRADE")
axes[0].scatter(df["studytime"],df["final_grade"],s=10,c=df["studytime"]) 
axes[1].boxplot(d,vert=True,patch_artist=True)
plt.show()
```
